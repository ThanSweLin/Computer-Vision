```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
new_img = cv2.flip(fix_img,-1)
```

```python
plt.imshow(new_img)
```

```python
new_img = fix_img.copy()
```

```python
cv2.rectangle(fix_img,pt1=(1500,800),pt2=(2225,1600),color=(255,0,0), thickness=10)
```

```python
plt.imshow(fix_img)
```

```python
tri_img = new_img.copy()

triangle_pts = np.array([(1750,800),(1350,1600),(2250,1600)])

cv2.drawContours(tri_img, [triangle_pts], 0, color=(0,0,255),thickness=10)
```

```python
plt.imshow(tri_img)
```

```python
tri_img_fill = new_img.copy()

triangle_pts = np.array([(1750,800),(1350,1600),(2250,1600)])

cv2.drawContours(tri_img_fill, [triangle_pts], 0, color=(0,0,255),thickness=-1)
```

```python
plt.imshow(tri_img_fill)
```
