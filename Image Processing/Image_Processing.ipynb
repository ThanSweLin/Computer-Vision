```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Task: Open and display the lion.jpg image that is located

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.imread('lion.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display_img(img)
```

### Task: Apply a binary threshold onto the image.

```python
img = cv2.imread('lion.jpg',0)
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
display_img(thresh1, cmap='gray')
```

### Task: Open the lion.jpg file and convert its colorspace to HSV and display the image.

```python
img  = cv2.imread('lion.jpg') 
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
display_img(img)
```

### Task: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10(0.01) and then use 2-D Convolution to blur the lion image(displayed in normal RGB)

```python
kernel = np.ones(shape=(4,4), dtype=np.float32)/10
```

```python
kernel
```

```python
img = cv2.imread('lion.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
result = cv2.filter2D(img,-1,kernel)
display_img(result)
```

### TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the lion image and then display the resulting gradient filtered version of the image

```python
img = cv2.imread('lion.jpg',0)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx,cmap='gray')
```

 ### Task: Plot the histograms for the RED, BLUE and GREEN channel of the lion image. Pay careful attention to the ordering of the channels.

```python
img = cv2.imread('lion.jpg')

color = ['b','g','r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr)
plt.title('My Histogram')
```
